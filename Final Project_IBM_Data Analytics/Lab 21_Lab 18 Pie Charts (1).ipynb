### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
db_series = (
    df["DatabaseWantToWorkWith"]
    .dropna()
    .astype(str)
    .str.split(";")
    .explode()
    .str.strip()
)

db_counts = db_series.value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(db_counts.values, labels=db_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Top 5 Desired Databases (DatabaseWantToWorkWith)")
plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
dev_series = (
    df["DevType"]
    .dropna()
    .astype(str)
    .str.split(";")
    .explode()
    .str.strip()
)

dev_counts = dev_series.value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(dev_counts.values, labels=dev_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Top 5 Developer Roles (DevType)")
plt.tight_layout()
plt.show()
```
